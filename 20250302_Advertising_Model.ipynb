```python
import pandas as pd # 데이터 처리 라이브러리로 판다스 사용
import numpy as np  # 수학적 함수들을 위한 넘파이 사용
import matplotlib.pyplot as plt # 시각화 작업을 위한 맷플롯립
from scipy import stats
import seaborn as sns # 맷플롭립에 기반한 데이터 시각화 라이브러리 사본
```

```python
%matplotlib inline
```

```python
url = 'https://github.com/books-by-chansung/low-code-ai/blob/main/advertising_2023.csv?raw=true'
```

```python
advertising_df = pd.read_csv(url, index_col=0)
```

```python
advertising_df.info()
```

```python
advertising_df.isnull().sum()
```

```python
advertising_df.head()
```

```python
advertising_df.describe()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(advertising_df.corr(), annot=True, vmin=0, vmax=1, cmap='viridis')
```

```python
advertising_df.plot(kind='scatter', x=['digital'], y='sales') # 산점도
```

```python
plt.figure(figsize=(18, 18))

# 가격 열(마지막 열)을 제외한 모든 열을 반복
for i, col in enumerate(advertising_df.columns[0:13]):
  plt.subplot(5, 3, i+1) # 열마다 세 개의 그래프
  x = advertising_df[col] # x축
  y = advertising_df['sales'] # y축
  plt.plot(x, y, 'o')

  # 회귀선 그리기
  plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1)) (np.unique(x)), color='red')
  plt.xlabel(col) # x축 레이블
  plt.ylabel('sales') # y축 레이블
```

```python
sns.displot(advertising_df, x="sales")
```

```python
lis = ['digital', 'newspaper', 'radio', 'TV']
plt.subplots(figsize=(15, 8))
index = 1
for i in lis:
  plt.subplot(2, 2, index)
  sns.distplot(advertising_df[i])
  index += 1
```
